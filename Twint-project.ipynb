```python
######### Twitter Scrapping By Oazed ##########

## Require twint
## Require asyncio
## Require Python 3.6

import twint
import pandas as pd
pd.options.display.max_columns = 100
import nest_asyncio
nest_asyncio.apply()

def twitter_scrape(_since, _until, _search, _out):
    import twint
    import nest_asyncio
    nest_asyncio.apply()
    x = twint.Config()
    x.Since = _since
    x.Until = _until
    x.Search = _search
#     x.Location = True
    x.Store_json =True
    x.Output = _out
    return twint.run.Search(x)

twitter_scrape("2022-01-01", "2022-01-24", "arbitrage", "../crypt-arbitrage.json")
```

```python
## For Analysis

tst= pd.read_json('../crypt-arbitrage.json', lines = True)
twit_en = tst.sort_values(by = "likes_count", ascending=False).reset_index()
twit_en = twit_en.drop(['index'],axis=1)
twit_en = twit_en.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))
twit_en.to_csv('../../crypto-arbitrage.csv',index=False)
```
